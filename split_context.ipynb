```python
import csv

## 資料預處理 取出q中只有一個句子
with open('data/Batch_answers - train_data (no-blank).csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    with open('output.csv', 'w', encoding='utf-8', newline='') as outfile:
        writer = csv.writer(outfile)
        for row in reader:
            if(len(row[1].split('.')) < 3):
                writer.writerow([row[0], row[1], row[4]])
```

```python
with open('output.csv', 'r', encoding='utf-8') as infile:
        reader = csv.reader(infile)
        nowID = -1
        numOfCases = 0
        correctCases = 0
        matchCase = 0
        for row in reader:
            if(nowID != row[0]):
                if(matchCase == 0):
                    print('not match ID: ', nowID)
                nowID = row[0]
                numOfCases += 1
                matchCase = 0

            if(matchCase):
                continue
            else:
                if(row[1] == row[2]):
                    correctCases += 1
                    matchCase = 1
        
        print("numOfCases:", numOfCases, "correctCases: ", correctCases, "accuracy: ", correctCases/numOfCases)
```
